```python
# Python SQL toolkit and Object Relational Mapper
import pandas as pd
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float
```

```python
Base = declarative_base()
engine = create_engine("sqlite:///heatmap.sqlite")
```

```python
jobsfile = "jobs.csv"
jobs_df = pd.read_csv(jobsfile)

citiesfile = "cities.csv"
cities_df = pd.read_csv(citiesfile)

recreationfile = "recreation.csv"
recreation_df = pd.read_csv(recreationfile)

housesfile = "houses.csv"
houses_df = pd.read_csv(housesfile)
  

# Define a cities class
class Cities(Base):
   __tablename__="cities"
   
   id = Column(Integer, primary_key=True)
   city = Column(String)
   state = Column(String)
   lat = Column(Float)
   long = Column(Float)

# Define a jobs class
class Jobs(Base):
   __tablename__="jobs"
   
   id = Column(Integer, primary_key=True)   
   city = Column(String)
   lat = Column(Float)
   long = Column(Float)
   job_type = Column(String)
   job_count = Column(Integer)

# Define a recreation class
class Recreation(Base):
   __tablename__="recreation"

   id = Column(Integer, primary_key=True)
   city = Column(String)
   lat = Column(Float)
   long = Column(Float)
   bars = Column(Integer)
   chinese_rest = Column(Integer)
   ice_cream_shops = Column(Integer)
   bike_trails = Column(Integer)


# Define a houses class
class Houses(Base):
   __tablename__="houses"

   id = Column(Integer, primary_key=True)
   city = Column(String)
   lat = Column(Float)
   long = Column(Float)
   house_count = Column(Integer)
   median_price = Column(Integer)

__table_args__ = {'extend_existing': True}

Base.metadata.create_all(engine)

jobs_df.to_sql(name = 'jobs', con = engine, if_exists ='replace', index=False)
cities_df.to_sql(name = 'cities', con = engine, if_exists ='replace', index=False)
recreation_df.to_sql(name = 'recreation', con = engine, if_exists ='replace', index=False)
houses_df.to_sql(name = 'houses', con = engine, if_exists ='replace', index=False)

# Declare a Base_1 using `automap_base()`
Base_1 = automap_base()

# Use the Base class to reflect the database tables
Base_1.prepare(engine, reflect=True)

# Print all of the classes mapped to the Base
Base_1.classes.keys()

# Assign the Base_1 class to a variable called `Jobs`
Jobs = Base_1.classes.jobs

# Assign the Base_1 class to a variable called `Cities`
Cities = Base_1.classes.cities

# Assign the Base_1 class to a variable called `Recreation`
Recreation = Base_1.classes.recreation

# Assign the Base_1 class to a variable called `Houses`
Houses = Base_1.classes.houses


# Create a session
session = Session(engine)
```
